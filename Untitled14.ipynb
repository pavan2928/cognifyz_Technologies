```python
text_reviews = d1['Rating text'].value_counts()
print("Rating text are as follows: ")
print(text_reviews)

sentiment_mapping = {
    'Good': 'Positive',
    'Very Good': 'Positive',
    'Excellent': 'Positive',
    'Poor': 'Negative',
    'Not rated': 'Neutral',
    'Average': 'Neutral',
}

custom_palette = ("husl", len(text_reviews))
d1['Sentiment'] = d1['Rating text'].map(sentiment_mapping)
plt.figure(figsize=(10, 6))
bars = plt.bar(text_reviews.index, text_reviews.values,)

plt.xlabel('Rating Text')
plt.ylabel('Count')
plt.title('Distribution of Rating Text')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.ylim(0, text_reviews.max() + 100)

for bar, count in zip(bars, text_reviews.values):
    plt.text(bar.get_x() + bar.get_width() / 2, count + 5, str(count), ha='center', va='bottom')

plt.show()

d1['Sentiment']
```

```python
d1['Review Length'] = d1['Rating text'].str.len()
d1['Review Length']
average_review_length = d1.groupby('Sentiment')['Review Length'].mean()

average_review_length.plot(kind='bar')
plt.xlabel('Sentiment')
plt.ylabel('Average Review Length')
plt.title('Relationship Between Review Length and Sentiment')
plt.xticks(rotation=45)
plt.show()

print("Average Review Length by Sentiment:")
print(average_review_length)
```

```python
from scipy.stats import pearsonr
sentiment_mapping = {
    'Positive': 1,
    'Neutral': 0,
    'Negative': -1,
}
d1['Sentiment_numeric'] = d1['Sentiment'].map(sentiment_mapping)

correlation_coefficient, _ = pearsonr(d1['Review Length'], d1['Sentiment_numeric'])

print(f"Correlation Coefficient: {correlation_coefficient}")
```

```python
highest_votes = d1.nlargest(1, 'Votes')
lowest_votes = d1.nsmallest(1, 'Votes')

result_df = pd.DataFrame({
    'Restaurant with High Votes': highest_votes.values[0],
    'Restaurant with Low Votes': lowest_votes.values[0]
}, index=d1.columns)
d1
```

```python
rating_analysis = d1['Aggregate rating'].corr(d1['Votes'])
print("Correlation between Aggregate rating and Votes:", rating_analysis)
```

```python
corr_matrix = d1.select_dtypes(["int64", "float64"]).corr()
plt.figure(figsize=(12, 8))
(corr_matrix, annot==True, cmap=='YlGnBu', fmt==".2f")
plt.title("Correlation Heatmap for Numerical Columns")
plt.show()
```

```python
grouped_data = d1.groupby(['Price range', 'Has Online delivery']).size().unstack()

colors = ['green', 'red']
ax = grouped_data.plot(kind='bar', color=colors, width=0.8)
plt.xlabel('Price Range')
plt.ylabel('Count')
plt.title('Distribution of Online Delivery by Price Range')
plt.xticks(rotation=0)
plt.legend(title='Has Online delivery')
plt.show()
```

```python
grouped_data = d1.groupby(['Price range', 'Has Table booking']).size().unstack()

colors = ['blue', 'yellow']
ax = grouped_data.plot(kind='bar', color=colors, width=0.8)
plt.xlabel('Price Range')
plt.ylabel('Count')
plt.title('Distribution of Table booking by Price Range')
plt.xticks(rotation=0)
plt.legend(title='Has Table booking')
plt.show()
```

```python
grouped_data = d1.groupby('Price range').agg(
    Online_delivery_percentage=('Has Online delivery', lambda x: (x == 'Yes').mean() * 100),
    Table_booking_percentage=('Has Table booking', lambda x: (x == 'Yes').mean() * 100),
)

# Plot
ax = grouped_data.plot(kind='bar', figsize=(9, 8))
plt.xlabel('Price Range')
plt.ylabel('Percentage')
plt.title('Percentage of Restaurants Offering Services by Price Range')
plt.xticks(rotation=0)
plt.legend(title='Service')
plt.show()
```

```python
from scipy.stats import chi2_contingency

price_online_contingency = pd.crosstab(d1['Price range'], d1['Has Online delivery'])

chi2, p, _, _ = chi2_contingency(price_online_contingency)
alpha = 0.05
if p < alpha:
    print("There is a statistically significant association between Price range and Online delivery.")
else:
    print("There is no statistically significant association between Price range and Online delivery.")

price_table_booking_contingency = pd.crosstab(d1['Price range'], d1['Has Table booking'])
chi2, p, _, _ = chi2_contingency(price_table_booking_contingency)

if p < alpha:
    print("There is a statistically significant association between Price range and Table booking.")
else:
    print("There is no statistically significant association between Price range and Table booking.")
```
